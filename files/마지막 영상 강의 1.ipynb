http 통신의 자동화가 가능하다.

```python
# post 요청 (민감한거 아이디 패스워드 같은)

data = {
    'id' : 'sdf'
    'pwd': '1224214w'
}
requests.post(url, data= data)
resp.text
```

```python
url = 'https://mk.co.kr/news/business/view/2019/06/432349/'
# HTTP header 이용하기

headers = {
    'user-agent': ''
}
requests.post(url, headers= headers)
resp.text
```

```python
url = 'https://mk.co.kr/news/business/view/2019/06/432349/'
resp = requests.get(url)
if resp.status_code == 200:
    print(resp.text)
else:
    print('err')
```
