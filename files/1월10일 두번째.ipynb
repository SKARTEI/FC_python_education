# filter, map, reduce

```python
#filter(함수, 리스트)
nums = [1,2,3,54,56,7]
def even(n):
    return n % 2 == 0

list(filter(even, nums))
# 리스트 함수 사용 필수!!!!!!!!!
# 리스트에서 조건 함수를 만족하는 것들만 추려낼 수 있음
```

```python
# map()
# 주어진 리스트에서 새로운 리스트로

nums = [1,2,3,54,56,7]
#map(함수, nums)
print(list(map(lambda n:n**2, nums)))
print(list(map(even, nums)))

#위는 lambda 간단하게, 밑에는 def로 정의 한 함수 사용
```
